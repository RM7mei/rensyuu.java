```python
def culsco2(points,tiled,width,height):
    tl=cir2(tiled,width,height)
    score=[0,0]
    for i in range(height):
        for j in range(width):
            if tl[i][j]==4:
                score[1]+=abs(points[i][j])
            if tl[i][j]==9:
                score[0]+=points[i][j]
    return score
```

```python
def move_pos_act2(agent,act,width,height,tiled):
    if act == 1:#if UP_LEFT==1
        if agent["x"]-1<=0 or agent["y"]-1<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]-2]==2:
            return False
        if tiled[agent["y"]-2][agent["x"]-2]==0:
            tiled[agent["y"]-2][agent["x"]-2]=1
        agent["x"]=agent["x"]-1
        agent["y"]=agent["y"]-1
        return [tiled,agent]
    elif act == 2:#if UP==2
        if agent["y"]-1<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]-1]==2:
            return False
        if tiled[agent["y"]-2][agent["x"]-1]==0:
            tiled[agent["y"]-2][agent["x"]-1]=1
        agent["y"]=agent["y"]-1
        return [tiled,agent]
    elif act == 3:#if UP_RIGHT==3
        if agent["x"]+1>width or agent["y"]<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]]==2:
            return False
        if tiled[agent["y"]-2][agent["x"]]==0:
            tiled[agent["y"]-2][agent["x"]]=1
        agent["x"]=agent["x"]+1
        agent["y"]=agent["y"]-1
        return [agent,tiled]
    elif act == 4:#if LEFT==4
        if agent["x"]-1<=0:
            return False
        if tiled[agent["y"]-1][agent["x"]-2]==2:
            return False
        if tiled[agent["y"]-1][agent["x"]-2]==0:
            tiled[agent["y"]-1][agent["x"]-2]=1
        agent["x"]=agent["x"]-1
        return [agent,tiled]
    elif act == 5:#stay==5とするなら
        return [agent,tiled]
    elif act == 6:#if RIGHT==6
        if agent["x"]+1 >width:
            return False
        if tiled[agent["y"]-1][agent["x"]]==2:
            return False
        if tiled[agent["y"]-1][agent["x"]]==0:
            tiled[agent["y"]-1][agent["x"]]=1
        agent["x"]=agent["x"]+1
        return [agent,tiled]
    elif act == 7:#if DOWN_LEFT==7
        if agent["x"]-1<=0 or agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]-2]==2:
            return False
        if tiled[agent["y"]][agent["x"]-2]==0:
            tiled[agent["y"]][agent["x"]-2]=1
        agent["x"]=agent["x"]-1
        agent["y"]=agent["y"]+1
        return [agent,tiled]
    elif act == 8:#if DOWN==8
        if agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]-1]==2:
            return False
        if tiled[agent["y"]][agent["x"]-1]==0:
            tiled[agent["y"]][agent["x"]-1]=1
        agent["y"]=agent["y"]+1
        return [agent,tiled]
    elif act == 9:#if DOWN_RIGHT==9
        if agent["x"]+1>width or agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]]==2:
            return False
        if tiled[agent["y"]][agent["x"]-1]==0:
            tiled[agent["y"]][agent["x"]-1]=1
        agent["x"]=agent["x"]+1
        agent["y"]=agent["y"]+1
        return [agent,tiled]
    else:
        return False
```
