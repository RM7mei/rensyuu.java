```python
def move_pos_act(agents,num,act,width,height,tiled):#numはID-1で そうじゃないほうがやりやすかったら内部で変更
    num -=1
    if type(act) == 'int':
        if act == 1:#if UP_LEFT==1
            if agents[num]["x"]-1<=0 or agents[num]["y"]-1<=0:
                return False
            if tiled[agents[num]["y"]-2][agents[num]["x"]-2]==2:
                return False
            agents[num]["x"]=agents[num]["x"]-1
            agents[num]["y"]=agents[num]["y"]-1
            return agents
        if act == 2:#if UP==2
            if agents[num]["y"]-1<=0:
                return False
            if tiled[agents[num]["y"]-2][agents[num]["x"]-1]==2:
                return False
            agents[num]["y"]=agents[num]["y"]-1
            return agents
        if act == 3:#if UP_RIGHT==3
            if agents[num]["x"]+1>width or agents[num]["y"]<=0:
                return False
            if tiled[agents[num]["y"]-2][agents[num]["x"]]==2:
                return False
            agents[num]["x"]=agents[num]["x"]+1
            agents[num]["y"]=agents[num]["y"]-1
            return agents
        if act == 4:#if LEFT==4
            if agents[num]["x"]-1<=0:
                return False
            if tiled[agents[num]["y"]-1][agents[num]["x"]-2]==2:
                return False
            agents[num]["x"]=agents[num]["x"]-1
            return agents
        if act == 5:#stay==5とするなら
            return agents
        if act == 6:#if RIGHT==6
            if agents[num]["x"]+1 >width:
                return False
            if tiled[agents[num]["y"]-1][agents[num]["x"]]==2:
                return False
            agents[num]["x"]=agents[num]["x"]+1
            return agents
        if act == 7:#if DOWN_LEFT==7
            if agents[num]["x"]-1<=0 or agents[num]["y"]+1>height:
                return False
            if tiled[agents[num]["y"]][agents[num]["x"]-2]==2:
                return False
            agents[num]["x"]=agents[num]["x"]-1
            agents[num]["y"]=agents[num]["y"]+1
            return agents
        if act == 8:#if DOWN==8
            if agents[num]["y"]+1>height:
                return False
            if tiled[agents[num]["y"]][agents[num]["x"]-1]==2:
                return False
            agents[num]["y"]=agents[num]["y"]+1
            return agents
        if act == 9:#if DOWN_RIGHT==9
            if agents[num]["x"]+1>width or agents[num]["y"]+1>height:
                return False
            if tiled[agents[num]["y"]][agents[num]["x"]]==2:
                return False
            agents[num]["x"]=agents[num]["x"]+1
            agents[num]["y"]=agents[num]["y"]+1
            return agents
    elif type(act) == 'str':
        if act == "UP_LEFT":
            if agents[num]["x"]-1<=0 or agents[num]["y"]-1<=0:
                return False
            if tiled[agents[num]["y"]-2][agents[num]["x"]-2]==2:
                return False
            agents[num]["x"]=agents[num]["x"]-1
            agents[num]["y"]=agents[num]["y"]-1
            return agents
        if act == "UP":
            if agents[num]["y"]-1<=0:
                return False
            if tiled[agents[num]["y"]-2][agents[num]["x"]-1]==2:
                return False
            agents[num]["y"]=agents[num]["y"]-1
            return agents
        if act == "UP_RIGHT":
            if agents[num]["x"]+1>width or agents[num]["y"]<=0:
                return False
            if tiled[agents[num]["y"]-2][agents[num]["x"]]==2:
                return False
            agents[num]["x"]=agents[num]["x"]+1
            agents[num]["y"]=agents[num]["y"]-1
            return agents
        if act == "LEFT":
            if agents[num]["x"]-1<=0:
                return False
            if tiled[agents[num]["y"]-1][agents[num]["x"]-2]==2:
                return False
            agents[num]["x"]=agents[num]["x"]-1
            return agents
        if act == "stay" or act == "STAY":
            return agents
        if act == "RIGHT":
            if agents[num]["x"]+1 >width:
                return False
            if tiled[agents[num]["y"]-1][agents[num]["x"]]==2:
                return False
            agents[num]["x"]=agents[num]["x"]+1
            return agents
        if act == "DOWN_LEFT":
            if agents[num]["x"]-1<=0 or agents[num]["y"]+1>height:
                return False
            if tiled[agents[num]["y"]][agents[num]["x"]-2]==2:
                return False
            agents[num]["x"]=agents[num]["x"]-1
            agents[num]["y"]=agents[num]["y"]+1
            return agents
        if act == "DOWN":
            if agents[num]["y"]+1>height:
                return False
            if tiled[agents[num]["y"]][agents[num]["x"]-1]==2:
                return False
            agents[num]["y"]=agents[num]["y"]+1
            return agents
        if act == "DOWN_RIGHT":
            if agents[num]["x"]+1>width or agents[num]["y"]+1>height:
                return False
            if tiled[agents[num]["y"]][agents[num]["x"]]==2:
                return False
            agents[num]["x"]=agents[num]["x"]+1
            agents[num]["y"]=agents[num]["y"]+1
            return agents
    else:
        return False
```

```python
def move_pos_act2(agent,act,width,height,tiled):
    if act == 1:#if UP_LEFT==1
        if agent["x"]-1<=0 or agent["y"]-1<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]-2]==2:
            return False
        if tiled[agent["y"]-2][agent["x"]-2]==0:
            tiled[agent["y"]-2][agent["x"]-2]=1
        agent["x"]=agent["x"]-1
        agent["y"]=agent["y"]-1
        return [tiled,agent]
    elif act == 2:#if UP==2
        if agent["y"]-1<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]-1]==2:
            return False
        if tiled[agent["y"]-2][agent["x"]-1]==0:
            tiled[agent["y"]-2][agent["x"]-1]=1
        agent["y"]=agent["y"]-1
        return [tiled,agent]
    elif act == 3:#if UP_RIGHT==3
        if agent["x"]+1>width or agent["y"]<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]]==2:
            return False
        if tiled[agent["y"]-2][agent["x"]]==0:
            tiled[agent["y"]-2][agent["x"]]=1
        agent["x"]=agent["x"]+1
        agent["y"]=agent["y"]-1
        return [agent,tiled]
    elif act == 4:#if LEFT==4
        if agent["x"]-1<=0:
            return False
        if tiled[agent["y"]-1][agent["x"]-2]==2:
            return False
        if tiled[agent["y"]-1][agent["x"]-2]==0:
            tiled[agent["y"]-1][agent["x"]-2]=1
        agent["x"]=agent["x"]-1
        return [agent,tiled]
    elif act == 5:#stay==5とするなら
        return [agent,tiled]
    elif act == 6:#if RIGHT==6
        if agent["x"]+1 >width:
            return False
        if tiled[agent["y"]-1][agent["x"]]==2:
            return False
        if tiled[agent["y"]-1][agent["x"]]==0:
            tiled[agent["y"]-1][agent["x"]]=1
        agent["x"]=agent["x"]+1
        return [agent,tiled]
    elif act == 7:#if DOWN_LEFT==7
        if agent["x"]-1<=0 or agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]-2]==2:
            return False
        if tiled[agent["y"]][agent["x"]-2]==0:
            tiled[agent["y"]][agent["x"]-2]=1
        agent["x"]=agent["x"]-1
        agent["y"]=agent["y"]+1
        return [agent,tiled]
    elif act == 8:#if DOWN==8
        if agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]-1]==2:
            return False
        if tiled[agent["y"]][agent["x"]-1]==0:
            tiled[agent["y"]][agent["x"]-1]=1
        agent["y"]=agent["y"]+1
        return [agent,tiled]
    elif act == 9:#if DOWN_RIGHT==9
        if agent["x"]+1>width or agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]]==2:
            return False
        if tiled[agent["y"]][agent["x"]-1]==0:
            tiled[agent["y"]][agent["x"]-1]=1
        agent["x"]=agent["x"]+1
        agent["y"]=agent["y"]+1
        return [agent,tiled]
    else:
        return False
```

```python
def break_act(agent,act,width,height,tiled):
    if act == 1:#if UP_LEFT==1
        if agent["x"]-1<=0 or agent["y"]-1<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]-2]==0:
            return False
        breaked=tiled[agent["y"]-2][agent["x"]-2]
        tiled[agent["y"]-2][agent["x"]-2]=0
        return [tiled,breaked]
    if act == 2:#if UP==2
        if agent["y"]-1<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]-1]==0:
            return False
        breaked=tiled[agent["y"]-2][agent["x"]-1]
        tiled[agent["y"]-2][agent["x"]-1]=0
        return [tiled,breaked]
    if act == 3:#if UP_RIGHT==3
        if agent["x"]+1>width or agent["y"]-1<=0:
            return False
        if tiled[agent["y"]-2][agent["x"]]==0:
            return False
        breaked=tiled[agent["y"]-2][agent["x"]]
        tiled[agent["y"]-2][agent["x"]]=0
        return [tiled,breaked]
    if act == 4:#if LEFT==4
        if agent["x"]-1<=0:
            return False
        if tiled[agent["y"]-1][agent["x"]-2]==0:
            return False
        breaked=tiled[agent["y"]-1][agent["x"]-2]
        tiled[agent["y"]-1][agent["x"]-2]=0
        return [tiled,breaked]
    if act == 6:#if RIGHT==6
        if agent["x"]+1 >width:
            return False
        if tiled[agent["y"]-1][agent["x"]]==0:
            return False
        breaked=tiled[agent["y"]-1][agent["x"]]
        tiled[agent["y"]-1][agent["x"]]=0
        return [tiled,breaked]
    if act == 7:#if DOWN_LEFT==7
        if agent["x"]-1<=0 or agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]-2]==0:
            return False
        breaked=tiled[agent["y"]][agent["x"]-2]
        tiled[agent["y"]][agent["x"]-2]=0
        return [tiled,breaked]
    if act == 8:#if DOWN==8
        if agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]-1]==0:
            return False
        breaked=tiled[agent["y"]][agent["x"]-1]
        tiled[agent["y"]][agent["x"]-1]=0
        return [tiled,breaked]
    if act == 9:#if DOWN_RIGHT==9
        if agent["x"]+1>width or agent["y"]+1>height:
            return False
        if tiled[agent["y"]][agent["x"]]==0:
            return False
        breaked=tiled[agent["y"]][agent["x"]]
        tiled[agent["y"]][agent["x"]]=0
        return [tiled,breaked]
    return False
```
