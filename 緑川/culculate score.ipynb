```python
from math import fabs
def culsco(a,b):
    k=cir(b)
    score=0
    for i in range(len(k)):
        for j in range(len(k[i])):
            if k[i][j]==4:
                score+=fabs(a[i][j])
            if k[i][j]==9:
                score+=a[i][j]
    return score
```

```python
def cir(b):
    c=[]
    for i in range(len(b)):
        z=[]
        for n in range(len(b[i])):
            z.append(0)
        c.append(z)
    for i in range(len(b)):
        for n in range(len(b[i])):
            if b[i][n]==1:
                c[i][n]=9
                for k in range(len(b[i])):
                    if k+1>n:
                        break
                    elif b[i][n-k-1]==1:
                        break
                    c[i][n-k-1]+=1
                for k in range(len(b[i])):
                    if n+k+1>=len(b[i]):
                        break
                    elif b[i][n+k+1]==1:
                        break
                    c[i][n+k+1]+=1
                for k in range(len(b)):
                    if k+1>i:
                        break
                    elif b[i-k-1][n]==1:
                        break
                    c[i-k-1][n]+=1
                for k in range(len(b)):
                    if i+k+1>=len(b):
                        break
                    elif b[i+k+1][n]==1:
                        break
                    c[i+k+1][n]+=1
    for s in range(len(b)):
        for n in range(len(b)):
            for i in range(len(b[n])):
                if c[n][i]<4:
                    if n-1>=0:
                        if not c[n-1][i]==9:
                            c[n-1][i]=0
                    if i-1>=0:
                        if not c[n][i-1]==9:
                            c[n][i-1]=0
                    if n+1<len(b):
                        if not c[n+1][i]==9:
                            c[n+1][i]=0
                    if i+1<len(b[n]):
                        if not c[n][i+1]==9:
                            c[n][i+1]=0
    return c
```
